```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.billboard.com/charts/billboard-200/"
```

```python
res = requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text, 'lxml')
```

```python
songs = soup.select("li h3")
print(songs)
```

```python
top200 = [song.text[14:][0:-5] for song in songs]
print(top200)
```

```python
top20 = top200[0:20]
print(top20)
```

```python
i = 0
for song in top20:
    i = i + 1
    print(f"{i}) {song}")
```
